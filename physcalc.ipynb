```python
from physpy import Engine, Body, Plotter
import numpy as np
import matplotlib.pyplot as plt

body1 = Body(name = "Body 1", mass = 1e4, type = "point", position = np.array([0, -3, 0]), velocity = np.array([0, 0, 0]), acceleration = np.array([0, 0, 0]), force = np.array([0, 0, 0]), charge = 1e-3)
body2 = Body(name = "Body 2", mass = 2e4, type = "point", position = np.array([10, 0, 5]), velocity = np.array([0, 0, 0]), acceleration = np.array([0, 0, 0]), force = np.array([0, 0, 0]), charge = -2e-5)

engine = Engine(bodies = [body1, body2], physics = ["dynamics", "gravity"], n_dimensions = 3, t_max = 1000)
engine.run()
```
